```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_path = '/beau_3.png'
target_path = '/new_target.png'

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.imread(image_path))
axes[1].imshow(cv2.imread(target_path))
plt.show()
```

```python
#1. 
image = cv2.imread(image_path)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

target = cv2.imread(target_path)
target_gray = cv2.cvtColor(target, cv2.COLOR_BGR2GRAY)
target_gray = cv2.resize(target_gray, (200,240)) # ,  ,      
#   200*240px  (**)
# :      

#2.   
w, h = target_gray.shape[::-1]
#  shape (, )
# [::-1]  >> (, )
# ?     

#3.   !
#  : SQDIFF( ), CCORR(  ), CCOEFF()  X NORMED(   )
result = cv2.matchTemplate(image_gray, target_gray, cv2.TM_CCOEFF_NORMED)
# SQDIFF( )  
# CCORR(  )  
# TM_CCOEFF_NORMED :  
# NORMED(   )  (0-1  )

#  , ,  ,    
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
# min_val:    
# min_loc :     (   )

# max_val, max_loc :    ,   (x,y)   

#
top_left = max_loc #     
#top_left = (x, y) top_left[0] x, top_left[1] y
bottom_right = (top_left[0]+w, top_left[1]+h) #    +, +    
# (top_left[0]+w, top_left[1]+h) : x+w. y+h
matched = image.copy()
cv2.rectangle(matched, top_left, bottom_right, (0, 255, 0), 2)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(image)
axes[1].imshow(result)
axes[2].imshow(matched)
plt.show()
```
